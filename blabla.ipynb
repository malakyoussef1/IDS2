```python
df['year']=df['date'].apply(lambda x: int(x[:4]))
df.head(5)
```

```python

plt.figure(figsize=(15,9))
sb.heatmap(df.corr(), annot=True)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['year'],df['Close'], label='Close Value', color='blue')
plt.plot(df['year'],df['High'], label='High Value', color='red')
plt.plot(df['year'],df['Low'], label='Low Value', color='green')
plt.plot(df['year'],df['Open'], label='Open Value', color='purple')
plt.title('Time Series Plot of Close Price')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
